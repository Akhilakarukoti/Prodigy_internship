```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset (replace 'train.csv' with your actual file path)
df = pd.read_csv('tested.csv')

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())

# Step 1: Data Cleaning

# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Fill missing 'Age' values with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing 'Embarked' values with the most common value
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop 'Cabin' column due to many missing values
df.drop(columns=['Cabin'], inplace=True)
```

```python
# Step 2: Exploratory Data Analysis (EDA)

# Summary statistics
print("\nSummary statistics:")
print(df.describe())

# Bar plot for categorical features like 'Pclass', 'Sex', and 'Embarked'
plt.figure(figsize=(18, 5))

# Passenger Class distribution
plt.subplot(1, 3, 1)
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')

# Gender distribution
plt.subplot(1, 3, 2)
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')

# Embarkation port distribution
plt.subplot(1, 3, 3)
sns.countplot(x='Embarked', data=df)
plt.title('Embarked Distribution')

plt.tight_layout()
plt.show()

# Histogram for 'Age'
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Box plot for 'Fare' by 'Pclass'
plt.figure(figsize=(8, 5))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
# Step 3: Analyzing Relationships

# Survival rate by gender
plt.figure(figsize=(8, 5))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.show()

# Survival rate by passenger class
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# Survival rate by age group
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 100], labels=['Child', 'Teen', 'Adult', 'Senior', 'Elderly'])
plt.figure(figsize=(8, 5))
sns.barplot(x='Age_Group', y='Survived', data=df)
plt.title('Survival Rate by Age Group')
plt.ylabel('Survival Rate')
plt.show()
```
